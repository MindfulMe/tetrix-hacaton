```python
import pandas as pd
```

```python
data = pd.read_csv("../../data/wine.csv", delimiter=",")
```

```python
colomns = []
for i in data[:]:
    d =   {
            "field": i, # which is the field's name of data key 
            "title": i, # display as the table header's name
            "sortable": True,
            }
    colomns.append(d)
# colomns = colomns[1:]
```
